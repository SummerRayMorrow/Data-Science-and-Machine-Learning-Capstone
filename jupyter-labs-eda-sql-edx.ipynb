## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1: Display the names of the unique launch sites  in the space mission


```python
# Task 1: Display the names of the unique launch_site
%sql select distinct Launch_Site from SPACEXTBL
```

### Task 2: Display 5 records where launch sites begin with the string 'KSC'


```python
# Task 2: Display 5 records where launch sites begin with the string 'KSC'
%sql select Launch_Site from SPACEXTBL where Launch_Site like 'KSC%' limit 5
```

### Task 3: Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Task 3: Display the total payload mass carried by boosters launched by NASA (CRS)
%sql select sum(Payload_Mass) as Total_Payload_Mass from SPACEXTBL where Customer = 'NASA (CRS)'
```

### Task 4: Display average payload mass carried by booster version F9 v1.1


```python
# Task 4: Display average payload mass carried by booster version F9 v1.1
%sql select avg(Payload_Mass) as Average_Payload_Mass from SPACEXTBL where Booster_Version = 'F9 v1.1'
```

### Task 5: List the date where the first successful landing outcome in drone ship was acheived.

*Hint:Use min function*


```python
# Task 5: List the date where the first successful landing outcome in drone ship was acheived.
%sql select MIN(TO_DATE(Date, 'DD/MM/YYYY')) from SPACEXTBL where Landing_Outcome = 'Success' and Landing_Location = 'Drone Ship'
```

### Task 6: List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
# Task 6: List the names of the boosters which have success in ground pad and have payload mass greater than 4000 but less than 6000
%sql select Booster from SPACEXTBL where Landing_Outcome = 'Success' and Landing_Location = 'Ground Pad' and Payload_Mass > 4000 and Payload_Mass < 6000
```

### Task 7: List the total number of successful and failure mission outcomes


```python
# Task 7: List the total number of successful and failure mission outcomes
%sql select COUNT(Mission_Outcome) as Total_Success from SPACEXTBL where Mission_Outcome = 'Success'
%sql select COUNT(Mission_Outcome) as Total_Failure from SPACEXTBL where Mission_Outcome != 'Success'
```

### Task 8: List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
# Task 8: List the names of the booster_versions which have carried the maximum payload mass. Use a subquery
%sql select DISTINCT Booster_Version from SPACEXTBL where Payload_Mass = (SELECT MAX(Payload_Mass) from SPACEXTBL)
```

### Task 9: List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017


```python
# Task 9: List the records which will display the month names, succesful landing_outcomes in ground pad, booster versions, launch_site for the months in year 2017
%sql select TO_CHAR(TO_DATE(Landing_Outcome_Date, 'DD/MM/YYYY'), 'Month'), Landing_Outcome, Booster_Version, Launch_Site from SPACEXTBL where Landing_Outcome like 'Success' and Landing_Location='Ground Pad' and strftime('%Y', Landing_Outcome_Date) = '2017'
```

### Task 10: Rank the  count of  successful landing_outcomes between the date 2010-06-04 and 2017-03-20 in descending order.


```python
# Task 10: Rank the  count of  successful landing_outcomes between the date 2010-06-04 and 2017-03-20 in descending order.
%sql select Landing_Outcome, Count(Landing_Outcome) as Successful_Landing_Count from SPACEXTBL where Landing_Outcome='Success' and Landing_Outcome_Date >= '2010-06-04' and Landing_Outcome_Date <= '2017-03-20' group by Landing_Outcome order by Successful_Landing_Count DESC
```
